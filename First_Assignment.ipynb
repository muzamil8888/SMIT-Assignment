#Topic 5: DICTIONARIES

```python
# Assignment 1: Student Info

student = {
    "name": "Ali",
    "age": 20,
    "grade": "A"
}

print(f"{student['name']} is {student['age']} years old and got grade {student['grade']}.")
```

```python
# Assignment 2: Update Dictionary

car = {"brand": "Toyota", "year": 2020}
car["color"] = "Red"
car["year"] = 2022

print("Updated Dictionary:", car)
```

```python
# Assignment 3: Loop Through Dictionary

countries = {
    "Pakistan": "Islamabad",
    "Turkey": "Ankara",
    "Japan": "Tokyo"
}

for country, capital in countries.items():
    print(f"{country} → {capital}")
```

```python
# Assignment 4: Dictionary of Prices

prices = {
    "apple": 1.5,
    "banana": 0.75,
    "orange": 1.2
}

product = input("Enter the product name (apple, banana, orange): ").lower()

if product in prices:
    print(f"The price of {product} is ${prices[product]}")
else:
    print("Sorry, that product is not available.")
```

```python
# Assignment 5: Nested Dictionary

students = {
    "Ali": {"age": 18, "grade": "A"},
    "Sara": {"age": 19, "grade": "B"}
}

for name, info in students.items():
    print(f"{name} got grade {info['grade']}.")
```
